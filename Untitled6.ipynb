```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import Dense, LSTM

```

**DATA LOADING & EXPLORATION**

```python
#loading dataset
file_path='/content/Samsung Dataset.csv'
data=pd.read_csv(file_path)
```

```python
#display first row
print(data.head())
```

```python
#displaying basic information about the dataset
print(data.info())
```

```python
#checking for missing values
print(data.isnull().sum())
```

**DATA PREPROCESSING**

```python
#converting data columns to date format
data['Date']=pd.to_datetime(data['Date'])
```

```python
#setting 'date' column as index
data.set_index('Date',inplace=True)
print(data.dtypes)
```

```python
print(data.describe())
```

**DATA ANALYSIS**

```python
#plotting the clossing price hsitory
plt.figure(figsize=(10,6))
plt.plot(data['Close'])
plt.title('Samsung Stock Price History')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.show()
```

```python
#correlation heatmep
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(),annot=True)
plt.show()
```

**FEATURE ENGINEERING**

```python
#creating features: Moving Average
data['SMA_20']= data['Close'].rolling(window=20).mean()
data['SMA_50']= data['Close'].rolling(window=50).mean()

#drop rows with NaN values
data.dropna(inplace=True)

#prepare the data for LSTM model
Scaler= MinMaxScaler(feature_range=(0,1))
scaled_data= Scaler.fit_transform(data[['Close']])
```

```python
# Creating a dataset with 60 timesteps and converting it to a numpy array
def create_dataset(dataset, time_step=60):
    x, y = [], []
    for i in range(len(dataset)-time_step-1):
        a = dataset[i:(i+time_step), 0]
        x.append(a)
        y.append(dataset[i + time_step, 0])
    return np.array(x), np.array(y)


time_step = 60
x, y = create_dataset(scaled_data, time_step)
```

```python
# Reshape input to be [samples, time steps, features] which is required for LSTM
x = x.reshape(x.shape[0], x.shape[1], 1)

#splitting the data into train & test data
train_size= int(len(x)*0.8)
test_size= len(x)-train_size
x_train, x_test= x[0:train_size], x[train_size:len(x)]
y_train, y_test= y[0:train_size], y[train_size:len(y)]
```

**MODEL BUILDING**

```python
#building LSTM model
model=Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(time_step,1)))
model.add(LSTM(50,return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')

#train the model
model.fit(x_train, y_train, batch_size=1, epochs=1)
```

**MODEL EVALUATION**

```python
#getting predicted stock values
train_predict= model.predict(x_train)
test_predict= model.predict(x_test)

#transform back to original form
train_predict= Scaler.inverse_transform(train_predict)
test_predict= Scaler.inverse_transform(test_predict)

#calculate RMSE
train_rmse= np.sqrt(mean_squared_error(y_train,train_predict[:, 0]))
test_rmse= np.sqrt(mean_squared_error(y_test,test_predict[:, 0]))

print(f'Train RMSE: {train_rmse}')
print(f'Test RMSE: {test_rmse}')
```

**FORECASTING FUTURE PRICES**

```python
#predicting the next 30 days
x_input= test_predict[-60:].reshape(1,-1)
temp_input = list(x_input)
temp_input= temp_input[0].tolist()

lst_output = []
n_steps=60
i = 0
while(i<30):
  if len(temp_input)>60:
    x_input=np.array(temp_input[1:])
    x_input=x_input.reshape(1,-1)
    x_input = x_input.reshape((1, n_steps, 1))
    yhat = model.predict(x_input, verbose=0)
    temp_input.extend(yhat[0].tolist())
    temp_input=temp_input[1:]
    lst_output.extend(yhat.tolist())
    i=i+1
  else:
    x_input = x_input.reshape((1, n_steps,1))
    yhat = model.predict(x_input, verbose=0)
    temp_input.extend(yhat[0].tolist())
    lst_output.extend(yhat.tolist())
    i=i+1
```

```python
#convert lst_output to a numpy array and reshape it to 2D
lst_output= np.array(lst_output).reshape(-1, 1)

#plotting the results
last_30_days= np.arange(1,31)
predicted_30_days= np.arange(31,61)

plt.figure(figsize=(14,7))
plt.plot(last_30_days, Scaler.inverse_transform(scaled_data[-30:]), label='Actual price')
plt.plot(predicted_30_days, Scaler.inverse_transform(lst_output), label='Predicted price')
plt.xlabel('Days')
plt.ylabel('Close Price(KRW)')
plt.title('Samsung Electronics Stock Prediction For Next 30 Days')
plt.legend()
plt.show()

```
